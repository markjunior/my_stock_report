```python
import pandas as pd
def ReadCurrentProfile():
    input_data = pd.read_excel("~/stock.xlsx", index_col=0)
    d = {'number': input_data['number'], 'unit_cost': input_data['unit_cost']}
    df = pd.DataFrame(data=d)
    return df;
    
ReadCurrentProfile().index
```

```python
def ReadLastestInfo():
    df = ReadCurrentProfile()
    latest_price = []
    for code in df.index:
        cur_company = yf.Ticker(code)
        hist = cur_company.history(period = "1d")
        latest_price.append(hist['Close'][-1])
    df['latest_price'] = latest_price
    return df
ReadLastestInfo()
    
```

```python
def GetSummaryDf():
    df = ReadLastestInfo();
    df['cost'] = df['unit_cost']*df['number']
    df['value'] = df['latest_price'] * df['number']
    df['revenue'] = df['value'] - df['cost']
    return df
    
GetSummaryDf()
```

```python
def get_current_timestamp():
    now = datetime.now()
    return now.strftime("%Y%m%d%H%M%S");


def output_to_csv(df):
#     if not os.path.exists('~/my_stock'):
#         os.makedirs('~/my_stock')
    file_name = "~/my_stock/summary_" + get_current_timestamp() + ".csv"
    df.to_csv(file_name, index=True)

    
    
output_to_csv(GetSummaryDf())
```

```python
def draw_pie(df, type):
    sizes = list(df[type])
    fig1, ax1 = plt.subplots(figsize=(20, 10))
    ax1.pie(sizes, labels=df.index, autopct='%1.1f%%',
            shadow=False, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.show()
    
draw_pie(GetSummaryDf(), 'cost')
```
